```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn import tree

import warnings
warnings.simplefilter('ignore')
```

```python
#reading data set to data frame
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#Storing features in X and target in y variables
```

```python
x = df[['SepalLengthCm', 'SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df[['Species']]
```

```python
#printing first 5 records
```

```python
print(x[:5])
```

```python
print(y[:5])
```

```python
#Performing train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=3)
```

```python
#Training the decision tree classifier
```

```python
#creating instance of decision tree
Tree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
Tree
```

```python
#training decision tree with train dataset
Tree.fit(X_train,y_train)
```

```python
#Visualization
```

```python
# make an instance of the model
fn = df.columns[1:5]
cn = df['Species'].unique().tolist()
Tree.fit(x,y)
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi = 300)

tree.plot_tree(Tree,
              feature_names=fn,
              class_names=cn,
              filled = True);
```

```python
#predicting values
```

```python
#making prediction after passing a single record
```

```python
X_new = [[6.7,3.0,6.4,2.3]]
```

```python
pred = Tree.predict(X_new)
pred
```
